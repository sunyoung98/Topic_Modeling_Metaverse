```python
import re
import pandas as pd
from tqdm import tqdm
from konlpy.tag import Okt
from collections import Counter
```

```python
zepeto=pd.read_csv("zepeto_review_test_ko.csv")
```

```python
zepeto = zepeto[zepeto['content'].notna()]
zepeto.isnull().sum()
```

```python
zepeto=zepeto.drop_duplicates()
zepeto.duplicated().sum()
```

```python
zepeto = zepeto.reset_index(drop=True)
```

```python
def extract_word(text):
    hangul = re.compile('[^가-힣]') 
    result = hangul.sub(' ', text) 
    return result
```

```python
zepeto['content'] = zepeto['content'].apply(lambda x:extract_word(x))
```

```python
from eunjeon import Mecab
import sys
import os
```

```python
m = Mecab('C:/Users/tjsdu/OneDrive/바탕 화면/데이터마이닝/mecab')
def tokenize_kn(text):  # 한글 형태소분석
    return " ".join([tok[0] for tok in m.pos(text)])
```

```python
zepeto['content_phrases']=""
tokenize_kn(zepeto['content'][0])
```

```python
zepeto['content_phrases']=zepeto['content'].apply(tokenize_kn)
zepeto['content_nouns']=zepeto['content_phrases'].apply(mecab.nouns).
```

```python
zepeto.to_csv('zepeto_phrases.csv',sep=',',na_rep='NaN',index=True)
```
