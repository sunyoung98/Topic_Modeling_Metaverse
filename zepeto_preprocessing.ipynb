```python
import re
import pandas as pd
from tqdm import tqdm
from konlpy.tag import Okt
from collections import Counter
```

```python
zepeto=pd.read_csv("zepeto_review_test_ko.csv")
```

```python
zepeto = zepeto[zepeto['content'].notna()]
zepeto.isnull().sum()
```

```python
zepeto=zepeto.drop_duplicates()
zepeto.duplicated().sum()
```

```python
zepeto = zepeto.reset_index(drop=True)
```

```python
def extract_word(text):
    hangul = re.compile('[^가-힣]') 
    result = hangul.sub(' ', text) 
    return result
```

```python
zepeto['content'] = zepeto['content'].apply(lambda x:extract_word(x))
```
